```python
import matplotlib.pyplot as plt
import numpy as np

frequency = np.linspace(405, 790, 80)
RefractiveIndex = (1 + ((1 / (1.731 - 0.261*((frequency / 1000)**2))) ** 0.5)) ** 0.5
Theta1 = np.arcsin(((4-RefractiveIndex**2)/3)**0.5)
Theta2 = np.arcsin(((9-RefractiveIndex**2)/8)**0.5)
Theta3 = np.pi/2
Phi1 = np.arcsin(np.sin(Theta1)/RefractiveIndex)
Phi2 = np.arcsin(np.sin(Theta2)/RefractiveIndex)
Phi3 = np.arcsin(np.sin(Theta3)/RefractiveIndex)
Phi1, Phi2, Phi3 = np.rad2deg(Phi1), np.rad2deg(Phi2), np.rad2deg(Phi3)

#These values are arbitrary, I just read of the graph given and these are roughly what they were
colors = []
for f in frequency:
    if f < 475:
        colors.append('red')
    elif f < 510:
        colors.append('orange')
    elif f < 530:
        colors.append('yellow')
    elif f < 600:
        colors.append('green')
    elif f < 620:
        colors.append('cyan')
    elif f < 675:
        colors.append('blue')
    else:
        colors.append('purple')

plt.scatter(frequency, Phi1, c=colors, s=12)
plt.plot(frequency, Phi1, color='blue')
plt.scatter(frequency, Phi2, c=colors, s=12)
plt.plot(frequency, Phi2, color='red')
plt.plot(frequency, Phi3, color='black')

plt.title("Refraction angle of single and double rainbows")
plt.xlabel('Frequency / THz')
plt.ylabel('$\\phi$ / deg')
plt.grid(True, alpha=0.5)
```
