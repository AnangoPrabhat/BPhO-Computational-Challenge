```python
# Use inline matplotlib plotting
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from random import randint
from time import time
import ipywidgets as widgets
from ipywidgets import interact

# --- Image Selection ---
# (This is as in your original code.)
choice = input("Tall (T) or Wide (W): ")
while choice.upper() not in ["T", "W"]:
    choice = input("Tall (T) or Wide (W): ")

if choice.upper() == "T":
    if randint(1, 22) == 7:
        image_file = "Tall2.jpg"  # Thales: 80 wide, 134 high
    else:
        image_file = "Tall1.jpg"  # Ella: 80 wide, 155 high
else:
    image_file = "aWide.jpg"      # 164 wide, 80 high

start_time = time()
image = plt.imread(image_file)
H, W = image.shape[0], image.shape[1]


def compute_S(offset_x, offset_y):
    """
    Compute a square canvas side length S (in pixels) that is as small as possible
    while ensuring that both copies of the image (with the given horizontal and vertical offsets)
    will be fully contained.

    Based on the original scale factor of 4 and then adding margin proportional to the absolute offset.
    """
    base = 4 * max(W, H)
    req_x = 4 * ((W - 1) + (offset_x))
    req_y = H + 2 * abs(offset_y)
    return int(max(base, req_x, req_y))


def create_canvas(S, offset_x, offset_y):
    """
    Create a square white canvas of size S x S and draw two copies of the image.

    The image is drawn with its "middle" at the point:
        (S // 2 + offset_x, S // 2 + offset_y).

    One copy is drawn normally on the right and the other copy is mirrored on the left.
    """
    canvas = np.full((S, S, image.shape[2]), 255, dtype=np.uint8)

    center_x = S // 2
    center_y = S // 2 + offset_y

    for y in range(H):
        for x in range(W):
            colour = image[y, x]
            # Right copy (normal)
            old_x = center_x + S // 4 + x + offset_x
            old_y = center_y - H // 2 + y
            new_x = center_x - S // 4 - x - offset_x
            new_y = center_y - H // 2 + y

            if 0 <= old_x < S and 0 <= old_y < S:
                canvas[old_y, old_x] = colour
            if 0 <= new_x < S and 0 <= new_y < S:
                canvas[new_y, new_x] = colour
    return canvas

def update_canvas(offset_x, offset_y):
    """
    Recompute the canvas (i.e. the positioning of the two image copies) based on the slider values
    and then redraw the plot.
    """
    S = compute_S(offset_x, offset_y)
    canvas = create_canvas(S, offset_x, offset_y)

    plt.figure(figsize=(6, 6))
    plt.imshow(canvas, extent=[-2, 2, -2, 2])
    plt.axvline(x=0, color='black', linestyle='--')
    plt.title(f"Offset X: {offset_x}, Offset Y: {offset_y}\nCanvas Size: {S}px")
    plt.show()

print("Initial processing time:", time() - start_time)

offset_range = max(W, H)

interact(update_canvas,
         offset_x=widgets.IntSlider(min=-offset_range, max=offset_range, step=1, value=0, description="Object X (px)"),
         offset_y=widgets.IntSlider(min=-offset_range, max=offset_range, step=1, value=0, description="Object Y (px)"));
```
