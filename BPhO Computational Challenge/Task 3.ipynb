```python
import matplotlib.pyplot as plt
import numpy as np

c = 3 * (10**8)
n = 1
y = 1
l = 2
x = np.linspace(0, l, 10000)

def GetTime(x):
    return (((x**2 + y**2)**0.5) / (c/n)) + ((((l-x)**2 + y**2)**0.5) / (c/n))

t = GetTime(x)
min = np.argmin(t)

plt.scatter(x[min], t[min] * 10**9, color='red', zorder=2)
plt.plot(x, t * 10**9, zorder=1)
plt.title(f"Minimum at x={round(x[min], 5)}")
plt.xlabel('x / m')
plt.ylabel('t / ns')
        
```
