```python
import matplotlib.pyplot as plt
import numpy as np

c = 3* 10**8
n1 = 1
n2 = 1.53
y = 1
Y = 1
L = 2
x = np.linspace(0,2,10000)

def GetTime(x):
    return (((x**2 + y**2)**0.5) / (c/n1)) + ((((L-x)**2 + y**2)**0.5) / (c/n2))

t = GetTime(x)
min = np.argmin(t)
theta1 = np.arctan(x[min] / y)
theta2 = np.arctan((L-x[min]) / Y)

plt.scatter(x[min], t[min] * 10**9, color='red', zorder=2)
plt.plot(x, t * 10**9, zorder=1)
plt.title(f"sin($\\theta_1$) / $c_1$: {round(np.sin(theta1) / (c/n1), 15)}.  sin($\\theta_2$) / $c_2$: {round(np.sin(theta2) / (c/n2), 15)}")
plt.xlabel('x / m')
plt.ylabel('t / ns')
```
